```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
df=pd.read_csv('McDonalds_financial_statements.csv')

df.head()
```

```python
# Extract Year and Dividend (stock split adjusted) columns
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Dividend (stock split adjusted) ($)'] = pd.to_numeric(df['Dividend (stock split adjusted) ($)'], errors='coerce')
data = df.dropna(subset=['Year', 'Dividend (stock split adjusted) ($)'])

X = data['Year'].values.reshape(-1, 1)  # Independent variable
y = data['Dividend (stock split adjusted) ($)'].values  # Dependent variable

# Linear Regression
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: Dividend vs Year')
plt.xlabel('Year')
plt.ylabel('Dividend (stock split adjusted) ($)')
plt.legend()
plt.grid()
plt.show()
```
